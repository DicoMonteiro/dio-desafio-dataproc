```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '***', '###!', '####', '****', '*****', 'NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE ou o estado == SP

filtro_classificacao = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro_uf = df.ocorrencia_uf == 'SP'

df.loc[filtro_classificacao | filtro_uf]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE ou classificação == INCIDENTE e o estado == SP

filtro_classificacao1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro_classificacao2 = df.ocorrencia_classificacao == 'INCIDENTE'
filtro_uf = df.ocorrencia_uf == 'SP'

df.loc[(filtro_classificacao1 | filtro_classificacao2) & filtro_uf]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE ou classificação == INCIDENTE e o estado == SP

filtro_classificacao = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro_uf = df.ocorrencia_uf == 'SP'

df.loc[filtro_classificacao & filtro_uf]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE ou classificação == INCIDENTE e o estado == SP

filtro_classificacao = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
filtro_uf = df.ocorrencia_uf == 'SP'

df.loc[filtro_classificacao & filtro_uf]
```

```python
# ocorrências cuja as cidades comecem com a letra 'C'

filtro_cidade = df.ocorrencia_cidade.str[0] == 'C'

df.loc[filtro_cidade]
```

```python
# ocorrências cuja as cidades terminam com a letra 'A'

filtro_cidade = df.ocorrencia_cidade.str[-1] == 'A'

df.loc[filtro_cidade]
```

```python
# ocorrências cuja as cidades terminam com os caracteres 'MA'

filtro_cidade = df.ocorrencia_cidade.str[-2:] == 'MA'

df.loc[filtro_cidade]
```

```python
# ocorrências cuja as cidades contém os caracteres 'MA' em qualquer parte do nome da cidade

filtro_cidade = df.ocorrencia_cidade.str.contains('MA')

df.loc[filtro_cidade]
```

```python
# ocorrências cuja as cidades contém os caracteres 'MA' ou 'AL' em qualquer parte do nome da cidade

filtro_cidade = df.ocorrencia_cidade.str.contains('MA|AL')

df.loc[filtro_cidade]
```

```python
# ocorrências do ano de 2015

filtro_ano = df.ocorrencia_dia.dt.year == 2015
df.loc[filtro_ano]
```

```python
# ocorrências do ano de 2015 e mês 12

filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
df.loc[filtro_ano & filtro_mes]
```

```python
# ocorrências do ano de 2015 e mês 12

filtro_ano_mes = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12)
df.loc[filtro_ano_mes]
```

```python
# ocorrências do ano de 2015 e mês 12 e o dia 08

filtro_ano_mes_dia = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & (df.ocorrencia_dia.dt.day == 8)
df.loc[filtro_ano_mes_dia]
```

```python
# ocorrências do ano de 2015 e mês 12 e o dia entre 03 e 08

filtro_ano_mes_dia = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & ((df.ocorrencia_dia.dt.day > 2) & (df.ocorrencia_dia.dt.day < 9))
df.loc[filtro_ano_mes_dia]
```

```python
# ocorrências contendo data (dia) e hora

df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```
